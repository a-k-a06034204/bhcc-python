## Step 5: Data cleaning - deleting duplicate entries

```python
''' Android '''

reviews_max = {}

# Grab the number of reviews per app:
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    # If we already saw this app name, but the number of reviews in the current entry is higher than what we have in our dictionary: 
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    
    # If we have not seen this app name before:
    elif name not in reviews_max:
        reviews_max[name] = n_reviews

print('Expected length:', len(android) - len(duplicate_apps))
print('Actual length:', len(reviews_max))

android_clean = []
already_added = []

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name) # make sure this is inside the if block
```

## Step 7: Data cleaning - removing apps with names of 3+ characters outside of ASCII range

```python
def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True

print(is_english('Docs To Go™ Free Office Suite'))
print(is_english('Instachat 😜'))


android_english = []
ios_english = []

# 
for app in android_clean:
    name = app[0]
    if is_english(name):
        android_english.append(app)
        
for app in ios:
    name = app[1]
    if is_english(name):
        ios_english.append(app)
        
explore_data(android_english, 0, 3, True)
print('\n')
explore_data(ios_english, 0, 3, True)
```

## Step 9: Data cleaning - Understand more of the data context 

As previously mentioned, the goal of this project is to determine the kinds of apps that are most likely to attract more users due to revenue being highly influenced by the number of people using certain apps.

In an attempt to minimize risks and overhead, this validation system for an app idea is made of 3 steps:
1.	Create a minimal Android version of the app, then add it to Google Play for users to enjoy.
2.	If the app has a positive response from its users, the app will be developed further.
3.	If the app is profitable after 6 months or less,an iOS version of the app will be created and added to the App Store for users to enjoy.
Because the end goal is to add the new app to both Google Play and the App Store for users to enjoy everywhere, analysis of app profiles that are successful on both markets is crucial. For example, a profile that works well for both markets may be a productivity app that makes use of gamification.


## Step 10: Data cleaning - frequency tables

```python
#isolating most common apps by genre

def freq_table(dataset, index):
    table = {}
    total = 0

    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1

    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage

    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

This section gets the average number of user ratings per app, per genre.
